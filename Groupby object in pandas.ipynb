## Group by --only on categorical column

```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
movies.head(2)
```

```python
geners = movies.groupby('Genre')
```

```python
# apply build in aggregation functions on group by object
geners.sum(numeric_only=True)
```

```python
geners.min()
```

```python
# find the top 3 geners by earning
movies.groupby('Genre').sum(numeric_only = True)['Gross'].sort_values(ascending = False).head(3)
```

```python
# method 2 is good bacause first it extract the column then sum  it so less time
movies.groupby('Genre')['Gross'].sum(numeric_only=True).sort_values(ascending=False).head(3)
```

```python
# find the gener with highest imdb rating
movies.groupby('Genre')['IMDB_Rating'].sum(numeric_only=True).sort_values(ascending=False)
```

```python
# find director with more popularity
movies.groupby('Director')['No_of_Votes'].sum(numeric_only=True).sort_values(ascending=False).head()
```

```python
# find the number of movies done by each actor
# movies['Star1'].value_counts()
# by using groupby
movies.groupby('Star1')['Runtime'].count().sort_values(ascending=False)
```

```python
# adding both the syntax
geners.agg({
    'Runtime': ['mean','min'],
    'IMDB_Rating':'mean',
    'No_of_Votes':['sum','max'],
    'Gross':'sum',
    'Metascore':'min'
})
```

```python
# looping on groups
# find the highest rating movie of each genre
df = pd.DataFrame(columns=movies.columns)
for group, data in geners:
    df = pd.concat([df, data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]])

df
```

```python
# find the number of movies starting with A
# split (apply) combine
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
geners.apply(foo)
```

```python
# find the ranking of each movie in the group according to IBMN score
def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
geners.apply(rank_movie)
```

```python
# find normalized IDMB rating group wise
# normalization farmula   X - Xmin / Xmax - Xmin
def normal(group):
    group['norm_rating'] = (group['IMDB_Rating']- group['IMDB_Rating'].min()) /( group['IMDB_Rating'].max()- group['IMDB_Rating'].min())
    return group
geners.apply(normal)
```

```python
# groupby on multiples columns
duo =movies.groupby(['Director','Star1'])
duo
# size
duo.size()
# get_group()
duo.get_group(('Peter Jackson','Elijah Wood'))
```

```python
# find the most earning actor,director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find actor,genre combo [the best in term of metascore(avg)]
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(3)
```

## Excercise

```python
ipl = pd.read_csv('deliveries.csv')
ipl.head(2)
```

```python
ipl.shape
```

```python
# find the top 5 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(5)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman')['batsman'].count().sort_values(ascending= False).head(1).
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
temp_df = ipl[ipl['over'] >15]
temp_df = temp_df[(temp_df['batsman_runs'] == 4 )|( temp_df['batsman_runs'] == 6)]

temp_df.groupby('batsman')['batsman'].count().sort_values(ascending= False).head(1).index[0]
```

```python
# find V Kohli's record against all teams
temp_df = ipl[ipl['batsman'] == 'V Kohli']

temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can return the highest score of any batsman
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('DA Warner')
```
